```python
data.columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm' , 'PetalWidthCm' , 'variety']
```

```python
m,n=data.shape
X=data.SepalLengthCm.iloc[0:50].values.reshape(50,1)
Y=data.SepalWidthCm.iloc[0:50].values.reshape(50,1)


Xnew=np.hstack((np.ones((50,1)),X))

np.random.seed(0)
theta=np.random.randn(1,2)
print('Theta:',theta)

iters=10000
J=np.zeros(iters)
learning_rate=0.001
```

```python
for i in range(iters):
    J[i]=(1/(2*m))*np.sum((np.dot(Xnew,theta.T)-Y)**2)
    theta[0,0]-=(learning_rate/m)*np.sum(np.dot(Xnew,theta.T)-Y)
    theta[0,1]-=(learning_rate/m)*np.sum((np.dot(Xnew,theta.T)-Y)*X)
plt.plot(np.arange(iters),J)
plt.xlabel('Number of iterations',color='orange')
plt.ylabel('Cost',color='orange')
plt.title('Cost vs iterations',color='green')
plt.show()
```

```python
print('Updated theta after Gradient Descent:',theta)
plt.scatter(X,Y,marker='.')
plt.plot(X,np.dot(Xnew,theta.T))
plt.xlabel('Sepal Length(in cm)')
plt.ylabel('Sepal Width(in cm)')
plt.title('Sepal length vs width of Iris-Setosa Species',color='green')
plt.show()
```

```python
A=data.PetalLengthCm.iloc[0:50].values.reshape(50,1)
B=data.PetalWidthCm.iloc[0:50].values.reshape(50,1)
Anew=np.hstack((np.ones((50,1)),X))
```

```python
np.random.seed(0)
theta=np.random.randn(1,2)
print('Theta:',theta)

iters=10000
J=np.zeros(iters)
learning_rate=0.001
```

```python
for i in range(iters):
    J[i]=(1/(2*m))*np.sum((np.dot(Anew,theta.T)-B)**2)
    theta[0,0]-=(learning_rate/m)*np.sum(np.dot(Anew,theta.T)-B)
    theta[0,1]-=(learning_rate/m)*np.sum((np.dot(Anew,theta.T)-B)*A)
plt.plot(np.arange(iters),J)
plt.xlabel('Number of iterations',color='orange')
plt.ylabel('Cost',color='orange')
plt.title('Cost vs iterations',color='green')
plt.show()
```

```python
print('Updated theta after Gradient Descent:',theta)
plt.scatter(A,B,marker='.')
plt.plot(A,np.dot(Anew,theta.T))
plt.xlabel('Petal Length(in cm)')
plt.ylabel('Petal Width(in cm)')
plt.title('Petal length vs width of Iris-Setosa Species',color='orange')
plt.show()
```
